## Why Use Matplotlib?

    -Core 2D plotting library in Python
    -Highly customizable and publication-quality
    -Integrates well with NumPy and Pandas
    -Supports line plots, bar charts, scatter plots, etc.
    -Works in Jupyter notebooks, scripts, and apps

```python
# Importing Libraries

import numpy as np 
import pandas as pd 

from matplotlib import pyplot as plt
```

## Ploting A Single Graph

```python
#Simple data
x = [10, 11, 12, 13, 14, 15, 16]
y = [100, 170, 200, 270, 300, 370, 400]

#Basic line plot

plt.plot(x,y)
plt.title("Basic Graph")
plt.xlabel('Sizd (m2)')
plt.ylabel('Price $')
plt.show()
```

## Plot 2 Graph On One Chart

```python
a = [10, 11, 12, 13, 14, 15, 16]
b = [100, 170, 200, 270, 300, 370, 400]
c = [90,110,130,150,170,190,210]

plt.plot(a,b)
plt.plot(a,c)
plt.title("Graph 1")
plt.xlabel("Size (m2)")
plt.ylabel("Price $")
plt.legend(['Line 1' , 'Line 2'])

plt.show()
```

## Styling with Colors, Linestyles, Markers

```python
plt.plot(a,b , color = 'g' , linestyle = '-.' , marker = 'o' , label = 'Line 1')
plt.plot(a,c , color = 'b' , linestyle = '--' , marker = '.' , label = 'Line 2')

plt.title("Graph After Styling")
plt.xlabel("Size (m2)")
plt.ylabel("Price $")
plt.legend()
plt.grid(True)
plt.show()
```

## Exploring Styles + Available Themes

```python
# Show available styles

print(plt.style.available)

# Use one

plt.style.use('ggplot')

#Re-plot with style

plt.plot(a, a, label='Line 1')
plt.plot(a, c, label='Line 2')
plt.title('Styled with ggplot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```

## Trying Another Style

```python
# Show available styles

print(plt.style.available)

# Use one

plt.style.use('_mpl-gallery')

#Re-plot with style

plt.plot(a, a, label='Line 1')
plt.plot(a, c, label='Line 2')
plt.title('Styled with _mpl-gallery')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```

# Importing Libraries & Reading Data

```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# Reading The CSV File Using Pandas

df = pd.read_csv('Cleaned_Orders.csv')
df.head()
```

## Convert Data Columns To DateTime



```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['ShippedData'] = pd.to_datetime(df['ShippedData'])

```

## Visualization 1 : Shipping Cost By County

```python
# Set the figure size for better readability
plt.figure(figsize=(6,4))
# Group data by country and calculate the average shipping cost per country
# Then sort the results for a cleaner horizontal bar chart

plt.style.use('_classic_test_patch')
df.groupby('ShipCountry')['ShippingCost'].mean().sort_values().plot(kind = 'barh' , color ='skyblue')

# Set Title For The Graph 
plt.title("Average Shipping Cost By Country")

# Label X , Y Axis
plt.xlabel("Average Shipping Cost")
plt.ylabel("Country")

# Show Grid Lines To Improve Readability
plt.grid(True)

# Automatically Adjust Layout To Prevent Label Cutoff
plt.tight_layout()
plt.show()
```

## Visualization 2: Delivery Days Distribution


```python
# Set the size of the figure
plt.figure(figsize=(10, 6))

# Plot a histogram of delivery days
# Drop any missing values to avoid errors
df["DeliveryDays"].dropna().plot(kind='hist' , bins=20 , color = '#649532ED' , edgecolor='k')

# Add a title to the chart
plt.title('Distribution Of Delivery Dayes')

# Label the X-axis and Y-axis
plt.xlabel("Delivery Dayes")
plt.ylabel('Frequency')

# Add gridlines for easier visual estimation
plt.grid(True)

# Adjust layout to avoid overlap or cut-off
plt.tight_layout()

# Render the plot
plt.show()
```

## Visualization 3: Delivery Status Count

```python
# Set the figure size (square shape is ideal for pie charts)
plt.figure(figsize=(5,5))

# Create a pie chart of the delivery status counts
df["DeliveryStatus"].value_counts().plot(kind = 'pie', colors = ["#FF0E05", "#CF4F1C","#0914B9A0"])

# Set the title of the plot
plt.title('Delivery Status BreakDown')

# Hide the y-axis label (not needed for pie charts)
plt.ylabel('')

# Adjust layout to make sure everything fits well
plt.tight_layout()

# Show the pie chart
plt.show()
```

## Visualization 4: Monthly Order Volume


```python
#Extract the month from the order date (format: YYYY-MM)
df["Monthes"] = df["OrderDate"].dt.to_period('M')

#Group the data by month and count the number of orders per month
Monthly_Order = df.groupby("Monthes").size()

#Set the figure size for better visibility
plt.figure(figsize=(12,6))

#Plot the monthly order volume as a line chart
Monthly_Order.plot(kind = 'line' , marker= 'o' , color = 'purple')

#Add a title to the plot
plt.title('Monthly Order Volume')

#Label the axes
plt.xlabel('Month')
plt.ylabel('Number Of Orders')

#Add gridlines for easier reading
plt.grid(True)

#Optimize layout spacing
plt.tight_layout()

#Display the chart
plt.show()
```
