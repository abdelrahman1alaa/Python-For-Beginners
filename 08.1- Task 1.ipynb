## Some of the orders are missing the unit_price.

**-Your task is to fill in each missing price using the average price for that same product, then Calculate total revenue across all orders.**

    -(Note: you need to have total price before calculating total revenue.)


**What is the total revenue after filling in the missing unit prices with the average price for each product?**


```python
df['unit_price'] = df['unit_price'].fillna(df['unit_price'].mean())
```

```python
df['unit_price'].head()
```

```python
df['total_price'] = df['quantity']*df['unit_price']
```

```python
df.head()
```

```python
df['total_revenue'] = df['total_price'].sum()
```

```python
df.head()
```

## Which product has the highest total quantity sold?



```python
print(df['quantity'].max())
```

```python
df.groupby("product")["quantity"].sum().idxmax()
```

## Which customer has the highest total spending across all orders?







```python
df.groupby("customer_name")["total_price"].sum().idxmax()
```
