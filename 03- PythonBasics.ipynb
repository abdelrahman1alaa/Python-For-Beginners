## While Loop

```python
#while loop -> execute a set of statements as long as a condition is true (i < 5)
i = 0
while i<5:
    print(i)
    i+=1
```

```python
#Draw a pyramid with *
i = 1
while i <= 5:
    print(i*"*")
    i+=1 #same as i = i + 1
```

```python
#break used to Terminates the loop even if the while condition is true

x = 10
while x >= 1:
    if x == 6:
        print("Terminates the loop")
        break
    print(x)
    x-=1
```

```python
#break used to Terminates the loop even if the while condition is true

x = 10
while x >= 1 :
    x-=1
    if x == 6:
        print("Terminates the loop")
        break
    print (x)
```

```python
#continue used to Skip only the current iteration then continue in while loop

i = 10
while i > 1:
    i-=1
    if i == 6:
        print("Skip this iteration")
        continue
    print(i)
```

## Task


Write a function called sum_list that takes a list of integers as input and returns the sum of all the numbers in the list.

Input Format: 
- A single line containing space-separated integers representing the elements of the list.
Output Format: 

- An integer representing the sum of all elements in the list.

Constraints: 

- The list will contain between 1 and 1000 integers.
- Each integer will be between -1000 and 1000.

Sample Input 1:
1 2 3 4 5

Sample Output 1:
15

Explanation 1:
1 + 2 + 3 + 4 + 5 = 15

Sample Input 2:
10 -5 7 3

Sample Output 2: 
15

Explanation 2:
10 + (-5) + 7 + 3 = 15

```python
def sum_list(numbers):
    return sum(numbers)

# Read input as space-separated integers
numbers = list(map(int, input().split()))

# Call the function and print the result
print(sum_list(numbers))
```
