## List

```python
#list -> collection of ordered and changeable elements

list1=[1,2,3,4,5]
print(list1)
```

```python
#slicing [start:end:steps] returns new list from the existing list

print(list1[0])         # ==> get the first element in the list

print(list1[0:3])      # ==> get elements from the first to element 3 "0,1,2"

print(list1[0:4:2])    # ==> get the elements from element 1 to element 3 and move with to steps 

print(list1[0:])       # ==> get the elements of the list from the first one to the end 
```

```python
print(list1[::2])      # ==> get all element with moving 2 steps 

print(list1[-1])       # ==> -1 means the last element

print(list1[-3])       # ==> -3 means the third element from back
```

```python
print([list1]*2)       # ==> get the list and repeat it twice "List of List "

print(list1[0:3]*2)    # ==> without { } that means no list of list , but mearge the element together
```

```python
# multiply each item by value 

print("The Original List : " , list1)
print("List After Multiply its Items By Values : " , [x*2 for x in list1])
```

```python
#append() is method that adds an element at the end of the list

Cars = ['BMW' , 'Ferrary' , 'F1' , 1990 , 1998 , 2024]
Cars.append("Maserati")
print(Cars)
```

```python
#insert() is method that adds an element at the specified position

Cars.insert(3,1999)
print(Cars)
```

```python
#remove() is method that removes the first item with the specified value

Cars.remove(1990)
print(Cars)
```

```python
# Printing The Type Of List

print(type(Cars))
```

## Dictionary

```python
#dictionary are used to store data values in key:value pairs
dict = {"name": "Ali" , 
        "age" : "20",
        "Job":"Data Engineer"}

print(dict)
```

```python
# print any key from the dictionary

print(dict["name"])
```

```python
# print the keys of the dictionary

print(dict.keys())
```

```python
# print the vlaues of the dictionary

print(dict.values())
```

```python
# print items of the dictionary

print(dict.items())
```
